```python
#importing libraries which is used in this model
import numpy as np
import pandas as pd
import ast
from sklearn.metrics import accuracy_score
```

```python
#read our dataset
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
#merging credits dataset with movies on the basis of title
movies=movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
#title 
#genre
#movies_id
#keyword
#cast
#crew
#overview
movies=movies[['movie_id','overview','title','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
#finding if there any null data is present
movies.isnull().sum()
```

```python
#droping the data with the null values
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
#["Action","Adventure","Fantasy","Science Fiction"]
```

```python
#function for only name  like action,adventure etc and apply it on genre and keyword
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
#Applied function for getting only top3 name
def change3(obj):
    L=[]
    count=0
    for i in ast.literal_eval(obj):
        if(count!=3):
            L.append(i['name'])
            count+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(change3)
```

```python
movies.head()
```

```python
#function for finding director name
def find_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
            
```

```python
movies['crew']=movies['crew'].apply(find_director)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#function to replace spaces b/w strings
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head()
```

```python
#creating tags by merging overview,genres,keywords,cast,crew
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
#creating new_df by merging movie-id,title and tags
new_df=movies[['movie_id','title','tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'][0]
```

```python
#converting string into lower case
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df
```

```python
#install nltk
pip install nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:11]
```

```python
#function for recommendation of movie name
def recommended(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:11]
    for i in movie_list:
        print(i)
```

```python
recommended('Avatar')
```
